```python
# create a function that would read the subscriber's last csv

def s3_list(prefix= ''):
    files = []
    resource = boto3.resource('s3')
    data_bucket = resource.Bucket('datascience.unbound.com')
    for obj in data_bucket.objects.filter(Prefix =prefix):
        files.append(obj.key)
    return files


def get_last_subs_csv_fname(folder ='subscriptions_csv_files/subscriptions'):
    timestamps = []
    for obj_name in s3_list(folder):
        
        try:
#get at time-stamp
            time_stamp = '20'+ obj_name.split('-20')[1].split('.')[0]
            ts = time_stamp.replace('_', ':')
            timestamps.append((pd.to_datetime(ts,yearfirst=True), obj_name))
        except:
            pass
    return max(timestamps)[1]
                              
    

#def s3_read_csv(key, index_col = index_col):
  #  csv_name = key
    
                          
                
```
